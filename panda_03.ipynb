### distribution plot


```python
# Histogram
sns.histplot(data = iris, x='petal_width', hue = 'species' ,bins=20)
plt.title("Distribution of petal_width")
plt.show
```

```python
sns.histplot(data =iris, x='sepal_length',bins = 20, hue ='species')
```

### barplot

```python
sns.barplot(data = iris, x='species',y ='sepal_length',palette ='Set2')
plt.title("Bar graph distribution of sepal_length based on species")
```
