### Dataframe for two dimensional dataset

```python
student_list = ["student1","student2","student3","student4","student5"]
roll_no = ["002","003","004","005","006"]
marks = [56.75,67.5,69.25,90.5,85.5]
```

```python
student_dict = {
    "Student Name": student_list,
    "Roll_no":roll_no,
    "Marks obtained":marks
}
```

```python
student_dict
```

```python
student_df = pd.DataFrame(student_dict)
student_df
```

```python
student_df.set_index("Roll_no",)
```

```python
student_df["Student Name"]
```

```python
student_df["Marks obtained"]
```

```python
student_df[["Student Name","Marks obtained"]]
```
