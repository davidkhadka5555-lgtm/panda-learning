### removing features(column)

```python
df = df.drop(columns=["PassengerId", "Survived", "Pclass", "Name"])
```

```python
df.head()
```

### Some feature engineering

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] +1
```

```python
df.head()
```

### Handeling missing values

```python
## For age
mean_age = df["Age"].mean()
mean_age = np.round(mean_age,2)
  
# mean_age

#for embarked
mode_embark = df["Embarked"].mode()[0]
mode_embark 
```

```python
df["Embarked"].value_counts
```

```python
mode = df["Embarked"].mode()[0]
mode
```

```python
df["Age"] = df["Age"].fillna(mean_age)
```

```python
df["Embarked"] = df["Embarked"].fillna(mode)
```

```python
df.isnull().sum()
```

```python
df["Has_Cabin"] = np.where(df["Cabin"].notna(), "Yes", "No")
```

```python
df.head()
```

```python
df = df.drop(columns=["Cabin"])
```

```python
df.head()
```

### Handaling catwgorical features(columns)

```python
df.head()
```

```python
df["Sex"] = df["Sex"].map({"male": 1, "female": 0})
```

```python
df.head()
```

```python
df["Has_Cabin"] = df["Has_Cabin"].map({"Yes": 1, "No": 0})
```

```python
df.head()
```

```python
df["Embarked"] = df["Embarked"].map({"S": 0, "C": 1, "Q": 2})
```
